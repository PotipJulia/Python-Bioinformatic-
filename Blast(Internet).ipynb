```python

from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML
from Bio import SearchIO
```

```python
seq=SeqIO.read(r"C:\Users\Alexander\Desktop\biopython\NC_005816.fna.txt", "fasta")
sequence_nucleotide=seq.seq
print(sequence_nucleotide)
```

```python
blast=NCBIWWW.qblast(program='blastn',database='nt',sequence=sequence_nucleotide)
with open('blast_result.xml','w') as file:
    file.write(blast.read())
blast.close()
```

```python
blast_qresult = SearchIO.read("blast_result.xml", "blast-xml")
print(blast_qresult)
```
