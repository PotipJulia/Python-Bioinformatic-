```python
import pandas as pd
import re
```

```python
pro_mouse_file=pd.read_csv(r'C:\Users\Alexander\Desktop\Gen_longevity\ortologues\1Mus_musculus_ENSMUSG00000000182\1Mus_musculus_ENSMUSG00000000182 (2).csv', sep=',')
pro_mouse_file
```

```python
typeValues=pro_mouse_file[pro_mouse_file['Type']=='1-to-1View Gene Tree'].index
drop=pro_mouse_file.drop(typeValues)
drop
```

```python
species_ortologue=drop[['Species','Orthologue']].drop_duplicates(subset='Species')
species_ortologue
```

```python
def get_orthologue(x):
    a,b = re.search(r'[A-z]+[0-9]{4}[0-9]+', x).span()
    return x[a:b]

ids = species_ortologue['Orthologue'].apply(lambda x: get_orthologue(x))
ids
```
